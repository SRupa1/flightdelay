```python
data['Org_Airport'].fillna(method = 'bfill' , inplace =True)
data['Dest_Airport'].fillna(method = 'bfill' , inplace =True)
```

```python
# Correct the date format to match 'DD-MM-YYYY'
data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y', errors='coerce')

# Check again for any remaining NaT values
print(data['Date'].isna().sum())  # This should ideally return 0 if all dates are now properly converted
```

```python
data = data.drop(columns=['Cancelled', 'CancellationCode'])
```

```python
data.drop(columns=['TailNum'], inplace=True)
```

```python
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
airline_delay = data.groupby('Airline')['ArrDelay'].mean()
sns.barplot(x=airline_delay.index, y=airline_delay.values)
plt.title('Average Arrival Delay by Airline')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Arrival Delay by Day of Week
plt.figure(figsize=(10, 6))
data.groupby('DayOfWeek')['ArrDelay'].mean().plot(kind='line', marker='o')
plt.title('Average Arrival Delay by Day of Week')
plt.xlabel('Day of Week (1 = Monday, 7 = Sunday)')
plt.ylabel('Average Arrival Delay')
plt.show()
```

```python
# Create two subsets: delayed and not delayed
delayed_flights = data[data['ArrDelay'] > 15]

# Plot histograms
plt.figure(figsize=(10, 6))

# Histogram for delayed flights
sns.histplot(delayed_flights['Distance'], color='red', label='Delayed Flights', kde=False, bins=30)

plt.title('Distance Distribution: Delayed vs. Non-Delayed Flights')
plt.xlabel('Distance (Miles)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
delay_columns = ['ArrDelay', 'DepDelay', 'CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']

# Calculate the correlation matrix
corr_matrix = data[delay_columns].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between ArrDelay and Other Delays')
plt.show()
```

```python
# Plot histogram for Taxi-Out times
plt.figure(figsize=(10, 6))
sns.histplot(data['TaxiOut'], color='blue', kde=False, bins=30)

plt.title('Distribution of Taxi-Out Times')
plt.xlabel('Taxi-Out Time (Minutes)')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
# Calculate total delay time by summing up delay types
data['total_delay'] = (data['CarrierDelay'] + data['WeatherDelay'] +
                     data['NASDelay'] + data['SecurityDelay'] + data['LateAircraftDelay'])

# Filter out rows with NaN in total_delay
df_delayed = data[~np.isnan(data['total_delay'])]

# Group by carrier and calculate the mean delay for each reason
carrier_group = df_delayed['CarrierDelay'].groupby(df_delayed['UniqueCarrier']).mean()
weather_group = df_delayed['WeatherDelay'].groupby(df_delayed['UniqueCarrier']).mean()
nas_group = df_delayed['NASDelay'].groupby(df_delayed['UniqueCarrier']).mean()
security_group = df_delayed['SecurityDelay'].groupby(df_delayed['UniqueCarrier']).mean()
late_group = df_delayed['LateAircraftDelay'].groupby(df_delayed['UniqueCarrier']).mean()

# Calculate positions for the stacked bars
w_bottom = carrier_group.values
n_bottom = w_bottom + weather_group.values
s_bottom = n_bottom + nas_group.values
l_bottom = s_bottom + security_group.values

# Get the carrier codes for x-axis labels
x = carrier_group.index.values

# Plot the stacked bar chart
fig, ax = plt.subplots(figsize=(12, 6))

ax.set_xticks(np.arange(len(x)))
ax.set_xticklabels(x, rotation=45)

# Plot each type of delay stacked on top of each other
ax.bar(np.arange(len(x)), carrier_group.values, align='center', label='Carrier Delay')
ax.bar(np.arange(len(x)), weather_group.values, align='center', bottom=w_bottom, label='Weather Delay')
ax.bar(np.arange(len(x)), nas_group.values, align='center', bottom=n_bottom, label='NAS Delay')
ax.bar(np.arange(len(x)), security_group.values, align='center', bottom=s_bottom, label='Security Delay')
ax.bar(np.arange(len(x)), late_group.values, align='center', bottom=l_bottom, label='Late Aircraft Delay')

# Set labels and title
ax.set_xlabel('Aircraft Carrier Code')
ax.set_ylabel('Average Delay in Minutes')
ax.set_title('Average Delay Reasons by Carrier')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Apply label encoding to categorical columns
data['UniqueCarrier'] = le.fit_transform(data['UniqueCarrier'])
data['Origin'] = le.fit_transform(data['Origin'])
data['Dest'] = le.fit_transform(data['Dest'])
data['Dest_Airport'] = le.fit_transform(data['Dest_Airport'])
data['Org_Airport'] = le.fit_transform(data['Org_Airport'])
 
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_features = ['ArrDelay', 'DepDelay', 'Distance', 'TaxiIn', 'TaxiOut', 'total_delay']  # Add other numerical features as needed
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```

```python
X = data.drop(columns=['ArrDelay','Airline'])  # Features
y = data['ArrDelay']  # Target variable
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error: {mae}')
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')  # Line for ideal predictions
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values')
plt.show()
```
